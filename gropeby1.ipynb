```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('weather_by_cities.csv')

# Define the custom function to group temperatures
def custom_temperature_groups(temperature):
    if 80 <= temperature <= 90:
        return 'Days between 80 and 90'
    elif 50 <= temperature <= 60:
        return 'Days between 50 and 60'
    else:
        return 'Days anything else'

# Apply the custom function to create the 'Temperature_Group' column
df['Temperature_Group'] = df['Temperature'].apply(custom_temperature_groups)

# Display the DataFrame with the new 'Temperature_Group' column
print(df)

# Other modifications or analysis on the DataFrame can be done here
# For example, you can calculate the average temperature for each group
average_temperatures = df.groupby('Temperature_Group')['Temperature'].mean()
print("\nAverage temperatures for each group:")
print(average_temperatures)

# You can also save the modified DataFrame back to a new CSV file
df.to_csv('weather_by_cities_modified.csv', index=False)
```
